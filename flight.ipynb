```python
df_test=final_df[final_df['Price'].isnull()]
```

```python
df_train=final_df[~final_df['Price'].isnull()]
```

```python
y=df_train['Price']
x=df_train.drop('Price',axis=1)
```

```python
x.shape
#y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x,y, test_size=0.33, random_state=42)
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
regressor=RandomForestRegressor()
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_features': [None, 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, 40, 50, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=param_dist,
            cv=5, n_iter=10,
            scoring = 'neg_mean_absolute_error',n_jobs = 3,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
random_cv.fit(X_train,y_train)
```

```python
k=random_cv.best_estimator_
```

```python
regressor.set_params(**k.get_params())
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print(f"root mean square error is:{rmse}")
```
